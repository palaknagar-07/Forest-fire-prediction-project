# Algerian Forest Fires Dataset - EDA & Feature Engineering,
This notebook walks through the process of loading, cleaning, and exploring the Algerian Forest Fires dataset. We'll perform feature engineering and visualize key insights.

```python
# importing required libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
# r\Read the data from csv
df = pd.read_csv("Algerian_forest_fires_dataset.csv", header= 1)
df.head()
```

```python
# check for null values
df.isnull().sum()
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
# Adding a column "Region" and encode it into 0 and 1 
df.loc[:122, "Region"] = 0
df.loc[122:,"Region"] = 1
```

```python
# float to int 
df["Region"] = df["Region"].astype(int)
df
```

```python
# Dropping the null values
df = df.dropna().reset_index(drop=True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.loc[[122]]
```

```python
df= df.drop(122).reset_index(drop=True)
```

```python
df.loc[[122]]
```

```python
# Removing the extra space from column name
df.columns = df.columns.str.strip()
```

```python
# Check the result of strip
df.columns
```

```python
# Type conversion for consistency and better understanding of data to model
df[["day", "month", "year", "Temperature", "RH", "Ws"]] = df[["day", "month", "year", "Temperature", "RH", "Ws"]].astype(int)
```

```python
df.info()
```

```python
# converting the remaining  features to float except 'Classes'
object = [features for features in df.columns if df[features].dtypes == 'O']
for i in object:
    if i != "Classes":
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
# Saving the cleaned data
df.to_csv("Forest_fire_updated_dataset.csv")
```

## Eploratory data analysis

```python
# Make a copy of df, for manipulation of data
df_copy = df
```

```python
# drop the irrelevent coloumns
df_copy = df_copy.drop(["day", "month", 'year'], axis=1)
df_copy
```

```python
# Encoding for consistency
df_copy["Classes"] = np.where(df_copy["Classes"].str.contains("not fire"), 0, 1)
```

```python
# Check the encoding
df_copy["Classes"].value_counts()
```

```python
plt.style.use("seaborn-v0_8")
df_copy.hist(bins=50, figsize=(20,15))
```

```python
# Pie chart of classes
percentage = df_copy["Classes"].value_counts(normalize=True)*100
classlables = ["Fire", "Not Fire"]
plt.figure(figsize=(12,6))
plt.pie(percentage, labels= classlables, autopct='%1.1f%%')
plt.title("Pie chart of classes")
```

```python
df_copy.corr()
```

```python
# Correlation visualization
sns.heatmap(df_copy.corr(), annot=True)
```

```python
# Encoding
df["Classes"] = np.where(df["Classes"].str.contains("not fire"), "Not fire","Fire")
```

```python
# Fire analysis sedi bel
dftemp = df.loc[df["Region"] == 1]
sns.countplot(x='month', hue= "Classes", data=dftemp)
plt.xlabel("Month", weight="bold")
plt.ylabel("No. of fire", weight="bold")
plt.title("Fire analysis in region of sedi-bel", weight="bold")
```

```python
# fire analysis of bejaia
dftemp = df.loc[df["Region"] == 0]
sns.countplot(x='month', hue= "Classes", data=dftemp)
plt.xlabel("Month", weight="bold")
plt.ylabel("No. of fire", weight="bold")
plt.title("Fire analysis in region of bejaia", weight="bold")
```
